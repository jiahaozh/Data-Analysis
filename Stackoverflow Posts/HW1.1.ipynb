````python
import requests
from datetime import datetime

start_time = 1420070400 # 01-01-2015 at 00:00:00
end_time   = 1420156800 # 01-02-2015 at 00:00:00

response = requests.get("https://api.stackexchange.com/2.2/questions?pagesize=100" +
                        "&fromdate=" + str(start_time) + "&todate=" + str(end_time) +
                        "&order=asc&sort=creation&site=stackoverflow")
print(response)
type(response)
````

Fill in the rest of the ```print_creation_dates_json()``` function that reads the response and prints the creation dates and whether the question was answered. Your solution should be in the form **(creation_date,is_answered)**. Notice that a JSON object is basically a dictionary. **(2 pts)**

````python
def print_creation_dates_is_answered_json(response):
    """
    Prints the creation_date of all the questions in the response.
    
    Parameters:
        response: Response object
    """
    for i in range(0,len(response['items'])):
        print("({},{})".format(datetime.utcfromtimestamp(response['items'][i]['creation_date']),(response['items'][i]['is_answered'])))
    
if __name__ == '__main__':
    print_creation_dates_is_answered_json(B)
````

## Step 3. Putting it all together

We are now ready to tackle our original problem. Write a function to measure the similarity of the top 100 users with the highest total answer scores. That is, you need to compute for each user the sum of the answer scores that he has received, for all answers that he has provided.

Compare the users based on the types of questions they answer. We will categorize a question by its first tag, so eventually the features of the user will comprise the first tag of each question that he has answered. You may choose to implement any one of the similarity/distance measures we discussed in class. **(2pts)**

Note that answers are posts with `PostTypeId=2`. The ID of the question in answer posts is the `ParentId`.

You may find the [sklearn.feature_extraction](http://scikit-learn.org/stable/modules/feature_extraction.html) module helpful.

````python
#def Similarity(file_path):
df = pd.read_csv('posts.csv')
A = df[df.Type == 'A']
#A
Q = df[df.Type == 'Q']
#Q
topuser=df[['OwnerUserId','Score']]
topuser=topuser.groupby('OwnerUserId').Score.sum().nlargest(100)
topuser=topuser.tolist()
#print(topuser)
ParentId_dict = dict()
for user in topuser:
    ParentId_dict[user] = A[(A['OwnerUserId'] == user)].ParentId.tolist()
#print(ParentId_dict)
Tag_dict = dict()
for user in topuser:
    for ParentId in ParentId_dict[user]:
        if ParentId not in Tag_dict:
            Tag_dict[ParentId] = Q[(Q['Id'] == ParentId)].Tags.tolist()
            #print(Tag_dict[ParentId])
#print(Tag_dict)
User_Tag_dict = dict()
for user in topuser:
    User_Tag_list = list()
    User_Question_list = ParentId_dict[user]
    for ParentId in User_Question_list:
        newtag = Tag_dict[ParentId]
        User_Tag_list.extend(newtag)
        User_Tag_list = list(set(User_Tag_list))
    User_Tag_dict[user] = User_Tag_list
print(User_Tag_dict)
````
